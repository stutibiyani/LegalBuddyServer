```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.read_csv('datasets/OPPBaseClassifierDataset.csv')
df = df.drop(['Id', 'ClauseID', 'JSON'], axis=1)
```

```python
import re
def clean(text):
    text = re.sub('r^([0-9]+\.)\\.?$', ' ', text)
    return text

df['Clause'] = df['Clause'].apply(lambda x: clean(x))
```

```python
df = df.drop_duplicates(['Clause', 'Category'])
df.shape
```

```python
df = df.groupby('Clause')['Category'].apply('| '.join).reset_index()
```

```python
df.head(6)
df.shape
```

```python
for x in range(len(df.index)): 
    df['Category'][x] = df['Category'][x].split('| ')
df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('Category')),
                          columns=mlb.classes_,
                          index=df.index))
```

```python
df.to_csv('datasets/BaseClassifierDataset.csv')
```
